```python
import pandas as pd
import numpy as np
```

```python
from pathlib import Path

files = {
    "users": Path("../data/users.csv"),
    "rooms": Path("../data/rooms.csv"),
    "typing_speeds": Path("../data/typing_speeds.csv"),
    "character_performance": Path("../data/character_performance.csv"),
}

def _resolve_path(p: Path) -> Path:
    if p.exists():
        return p
    alt = Path("/") / p
    if alt.exists():
        return alt
    raise FileNotFoundError(f"CSV not found: {p}")

df_users = pd.read_csv(_resolve_path(files["users"]))
df_room = pd.read_csv(_resolve_path(files["rooms"]))
df_typing_speeds = pd.read_csv(_resolve_path(files["typing_speeds"]))
df_character_performance = pd.read_csv(_resolve_path(files["character_performance"]))

print(df_users.shape, df_room.shape, df_typing_speeds.shape, df_character_performance.shape)
```

#set of dataType and Wants rows

```python
selectedRoom_cols = ['id','createdAt','textLength']
selectedtypingSpeed_cols = ['id','userId','roomId','wpm','correctword','duration']
selectedCharacterPerformance_cols = ['id','char','avgTimePerChar','maxTimePerChar','errorFrequency','typingSpeedId','userId']

set_room = df_room[selectedRoom_cols]
set_typing = df_typing_speeds[selectedtypingSpeed_cols]
set_char = df_character_performance[selectedCharacterPerformance_cols]

union_df = (
    df_room.add_suffix('_room')
    .merge(
        df_typing_speeds.add_suffix('_typing'),
        left_on='id_room',
        right_on='roomId_typing',
        how='inner'
    )
    .merge(
        df_character_performance.add_suffix('_char'),
        left_on='id_typing',
        right_on='typingSpeedId_char',
        how='inner'
    )
)

print("Final merged DataFrame shape:", union_df.shape) 

```
